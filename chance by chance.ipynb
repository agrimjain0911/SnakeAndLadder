```python
def RandomGenerator():
    import random
    num= random.randint(1,6)
    if num != 6:
        return num
    else:
        totalnum = 6
        
        for i in range(0,3):
            j = random.randint(1,6)
            if(j!=6):
                return totalnum + j
        return random.randint(1,6)
```

```python
def game(times,P1,P2):
    total_p1=0
    total_p2=0
    for i in range(1,times+1):
        for j in range(1,3):
            if j==1:
                num = RandomGenerator()
                total_p1=total_p1+num
                print(P1+" = " + str(num) )
                print(total_p1)
                if total_p2==total_p1:
                    total_p2=0
                    print(P2 +" has to start all over")
                    print()
            if j==2:
                num = RandomGenerator()
                total_p2=total_p2+num
                print(P2+" = " + str(num) )
                print (total_p2)
                print()
        if total_p1==total_p2:
            total_p1=0
            print(P1 +" has to start all over")
            print()
    print("Total count of " + P1 + " = " + str(total_p1))
    print("Total count of " + P2 + " = " + str(total_p2))
    print()
    if total_p1>total_p2:
        print(P1 + " won by " + str(total_p1-total_p2))
    elif total_p1==total_p2:
        print("It is a tie game")
    else:
        print(P2 + " won by " + str(total_p1+total_p2))   
```
